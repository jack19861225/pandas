```python
%pylab inline
import pandas
sephora_df = pandas.read_table('sephora.txt', header=0, names=['PROD', 'DOWNLOAD_DATE', 'PRICE'])
sephora_df
```

```python
def avg_price(price):
    price = price.replace('$', '').replace(' ', '').strip()
    avg_p = price.split("-")
    if len(avg_p) == 2:
        return (float(avg_p[0]) + float(avg_p[1]))/2.0
    elif len(avg_p) == 1:
        return float(avg_p[0])
    else:
        return float(-1)
```

```python
sephora_data = sephora_df
sephora_data['AVGPRICE'] = [avg_price(x) for x in sephora_data.PRICE]
sephora_data
```

```python
sephora_count = pandas.DataFrame(sephora_data, columns=['PROD', 'DOWNLOAD_DATE'])
sephora_plot = sephora_count.groupby(sephora_count.DOWNLOAD_DATE).count()
sephora_plot
```

```python
sephora_plot.plot(kind="bar", stacked=True, alpha=0.5)
```

```python
sephora_plot['PROD']
sephora_plot['PROD'].plot(kind="bar", stacked=True, alpha=0.5)
```

```python
sephora_buckets = sephora_data
sephora_buckets['<20'] = sephora_buckets['AVGPRICE'] < 20
sephora_buckets['20-40'] = (sephora_buckets['AVGPRICE'] >= 20) & (sephora_buckets['AVGPRICE'] < 40)
sephora_buckets['40-60'] = (sephora_buckets['AVGPRICE'] >= 40) & (sephora_buckets['AVGPRICE'] < 60)
sephora_buckets['>=60'] = sephora_buckets['AVGPRICE'] >= 60
sephora_buckets
```

```python
sephora_price_buckets = sephora_buckets.groupby(sephora_buckets.DOWNLOAD_DATE).sum()
```

```python
sephora_prices = pandas.DataFrame(sephora_price_buckets, columns=['<20', '20-40', '40-60', '>=60'])
```

```python
price_plot = sephora_prices.plot(kind="bar", stacked=True, alpha=0.8)
price_plot.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
sephora_prices.div(sephora_prices.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, alpha=0.8).legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
